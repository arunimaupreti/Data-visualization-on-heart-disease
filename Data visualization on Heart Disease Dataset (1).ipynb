```python
df = pd.read_csv('heart.csv')
```

```python
df.head()
```

#  Distribution plot for numerical column

```python
sns.displot(df['Cholesterol'], kde = True, rug = True)
```

# Pie charts for Categorical

```python
df.groupby(df['Sex']).size().plot(kind = 'pie', autopct = '%.1f')
```

```python
df.groupby(df['ChestPainType']).size().plot(kind = 'pie', autopct = '%.1f')
```

```python
df.groupby(df['RestingECG']).size().plot(kind = 'pie', autopct = '%.1f')
```

```python
df.groupby(df['ST_Slope']).size().plot(kind = 'pie', autopct = '%.1f')
```

```python
sns.violinplot(y =df['Age'], x = df['HeartDisease'])
```

```python
sns.violinplot(y =df['RestingBP'], x = df['HeartDisease'])
```

# JointPLot

```python
df.head()
```

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df)
```

```python
sns.jointplot(x = 'Age', y = 'Age', data = df)
```

```python
sns.jointplot(x = 'HeartDisease', y = 'Oldpeak', data = df) # No correlation
```

```python
sns.jointplot(x = 'Cholesterol', y = 'Oldpeak', data = df)
```

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df,kind = 'hex')
```

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df,kind = 'reg')   # if the age increases the max heart rate decreases
```

```python
sns.jointplot(x = 'Cholesterol', y = 'FastingBS', data = df,kind = 'reg') # No correlation
```

```python
sns.jointplot(x = 'HeartDisease', y = 'MaxHR', data = df,kind = 'reg')
```
